```python
# merge and join operation 
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                     'key2':[4,5,6,7,8],
                     'key3':[3,4,5,6,6]})
```

```python
data1
```

```python
data2 = pd.DataFrame({'key1':[1,2,3,4,24],
                     'key4':[32,23,4,33,44],
                     'key5':[23,56,54,23,22]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2) # merge on the based of indexlabel  and do inner merge 
# it is trying to do the inner merge operation
# trying to do things like intersaction .
```

```python
pd.merge(data1,data2,how='outer')
```

```python
pd.merge(data1,data2,how='left')
```

```python
pd.merge(data1,data2,how='right')
```

```python
pd.merge(data1,data2,how='cross') # it will take data1 and combine with everything one by one 
```

```python
data1 # took 0th index 1,4,3 combine with every data of data2
```

```python
# merge will try to find for the common column we can also give the column name purposly
```

```python
pd.merge(data1,data2, on = 'key1' , how = 'left')
```

```python
pd.merge(data1,data2, on = 'key2' , how = 'right')
```

```python
pd.merge(data1,data2 , how='left' , left_on = 'key2' , right_on = 'key4')
```
